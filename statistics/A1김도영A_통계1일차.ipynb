### 자동차 연비 Data set에서 model, driving 데이터에 대해   
- counts,percents, cumulative counts, cumulative percents 구하기 (p28) -data.set : mycars.csv

### model에 대해

```python
ds_mycars = pd.read_csv("/home/piai/POSCO_AI_AC/statistic/mycars.csv",engine="python")
ds_mycars.head()

df_model = ds_mycars[['model']]
df_model.head()
```

```python
count = df_model['model'].value_counts().sort_index()
count[:5]
```

```python
cumcnt = np.cumsum(count)
cumcnt[:5]
```

```python
percent = count/sum(count)*100
percent[:5]
```

```python
cumpct = np.cumsum(percent)
cumpct[:5]
```

```python
count_data = pd.DataFrame({'Count':count, 
                           'Percent':percent,
                           'Cumulative counts':cumcnt,
                           'Cumulative percents':cumpct})
count_data.columns.name = "model"
count_data
```

mycars.csv 데이터는 총 234대의 차량의 정보를 담고 있다. 모델을 기준으로 하여 counts, percents, cumulative counts, cumulative percents를 구하였을 때 가장 많은 모델은 caravan 2wd 이고 가장 적은 모델은 land cruiser wagon 4wd이다. 이 분류를 통해 이 데이터에 차량 모델 별 수량과 전체 중 몇 퍼센트를 차지하는지 알 수 있었다.
